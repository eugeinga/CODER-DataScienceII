* **Acidez fija:** La acidez fija es responsable del sabor 치cido del vino. Los vinos con una acidez fija en el rango r칤pico suelen tener un sabor fresco y equilibrado. Los valores m치s altos son m치s comunes en vinos blancos y espumosos.
* **Acidez vol치til:** Niveles altos de acidez vol치til pueden dar un sabor avinagrado al vino, lo cual generalmente es un defecto.
* **츼cido c칤trico:**	Es un 치cido que agrega frescura y un sabor afrutado al vino. Su presencia en peque침as cantidades es deseable.
* **Az칰car residual:**	Es el az칰car que queda en el vino despu칠s de la fermentaci칩n. Los vinos con menos az칰car residual son secos, mientras que los vinos con m치s az칰car residual son dulces.
* **Cloruros:**	Se refiere al contenido de sal en el vino, que puede influir en su sabor y cuerpo. Niveles altos pueden dar un sabor salado, lo que generalmente es indeseable.
* **Di칩xido de azufre libre:**	Es el di칩xido de azufre que no est치 ligado a otros compuestos en el vino y act칰a como un conservante, ayudando a prevenir la oxidaci칩n y el crecimiento microbiano. Los vinos blancos y rosados suelen requerir m치s di칩xido de azufre libre que los tintos.
* **Di칩xido de azufre total:**	Es la suma del di칩xido de azufre libre y el combinado. Un nivel adecuado es importante para la conservaci칩n del vino, pero niveles demasiado altos pueden causar un sabor fuerte y desagradable. Los valores m치ximos permitidos var칤an seg칰n la legislaci칩n de cada pa칤s, generalmente se permite hasta 200 mg/L para vinos tintos y hasta 350 mg/L para vinos blancos y rosados.
* **Densidad:**	Se refiere a la relaci칩n entre la masa y el volumen del vino. Est치 estrechamente relacionada con el contenido de az칰car y alcohol; un vino m치s dulce o con m치s alcohol generalmente tendr치 mayor densidad. Los vinos secos tienden a estar m치s cerca del rango inferior, mientras que los vinos dulces est치n en el extremo superior.
* **pH:**	Mide la acidez o alcalinidad del vino. Un pH bajo (com칰n en vinos blancos) indica alta acidez, lo que contribuye a la frescura y estabilidad del vino.
* **Sulfatos:**	Los sulfatos, como el sulfato de potasio, son aditivos que pueden usarse para estabilizar el vino y mejorar su claridad. En cantidades excesivas pueden influir negativamente en el sabor y la sensaci칩n en boca.
* **Alcohol:**	Indica el porcentaje de alcohol por volumen en el vino. Afecta el cuerpo, el sabor y la percepci칩n general del vino. Un vino con m치s alcohol puede parecer m치s c치lido y robusto. Los vinos de mesa suelen estar en el rango de 11% a 14%.
* **Calidad:**	Este es un valor subjetivo basado en una evaluaci칩n sensorial y puede variar ampliamente seg칰n el catador y el tipo de vino. Un vino de alta calidad generalmente estar치 en el rango de 7 a 10.

#### **餃 FEATURE ENGINEERING**
**(Generaci칩n de Nuevas Caracter칤sticas)**

Se crean nuevas variables a partir de las existentes a fin de aportar mayor informaci칩n al modelo de Machine Learning objetivo ayudando a capturar relaciones m치s complejas entre las caracter칤sticas del vino y mejorar su rendimiento.

1. Relaciones entre 츼cidos: Se sumar la acidez fija y la acidez vol치til para obtener una medida de la acidez total. $$ \text{Total Acidity} = \text{Fixed Acidity} + \text{Volatile Acidity} $$

```python
df['total acidity'] = df['fixed acidity'] + df['volatile acidity']
```

2. Proporciones de Azufre: Se calcula la proporci칩n de di칩xido de azufre libre respecto al total. $$ \text{Prop Sulful} = \frac{\text{Free Sulfur Dioxide}}{\text{Total Sulfur Dioxide}} $$

```python
df['prop sulful'] = df['free sulfur dioxide'] / df['total sulfur dioxide']
```

3. Densidad Ajustada ('ajust density'): Se ajusta la densidad del vino teniendo en cuenta el contenido de alcohol. $$ \text{density Ajust} = \frac{\text{Density}}{\text{Alcohol}} $$

```python
df['ajust density'] = df['density'] / df['alcohol']
```

4. 칈ndice de Dulzura: Se calcula la proporci칩n de az칰car residual respecto a la acidez total. $$ \text{Sweetness} = \frac{\text{Residual Sugar}}{\text{Total Acidity}} $$

```python
df['sweetness'] = df['residual sugar'] / df['total acidity']
```

5. 칈ndice de Sulfitos: Se calcula la proporci칩n de sulfatos respecto a los cloruros. $$ \text{sulphate index} = \frac{\text{Sulphates}}{\text{Chlorides}} $$

```python
df['sulphate index'] = df['sulphates'] / df['chlorides']
```

6. 칈ndice de Acidez: Se ajustar la acidez fija por el pH para obtener una medida m치s precisa de la acidez percibida. $$ \text{acidity index} = \frac{\text{Fixed Acidity}}{\text{pH}} $$

```python
df['acidity index'] = df['fixed acidity'] / df['pH']
```

**MEDIDAS DE ASIMETR칈A Y CURTOSIS**

Estas medidas informan sobre dos aspectos importantes de la forma de la distribuci칩n: su grado de asimetr칤a y su grado de homogeneidad.

Un valor de **asimetr칤a** (*skewness*) cercano a 0 (entre -0.5 y 0.5) indica que los datos est치n sim칠tricamente distribuidos alrededor de la media. Valores positivos indican una distribuci칩n sesgada a la derecha (cola larga a la derecha), mientras que valores negativos indican una distribuci칩n sesgada a la izquierda (cola larga a la izquierda)

Por su parte, la **curtosis** (*kurtosis*) mide la heterogeneidad de la distribuci칩n de los datos (o la concentraci칩n de los valores de la variable). La curtosis de Fisher, que es la que calcula 'scipy.stats.describe', est치 normalizada para que una distribuci칩n normal tenga una curtosis de 0. Valores positivos indican mayor concentraci칩n de los datos en torno a la media; es decir, una distribuci칩n con colas m치s pesadas (m치s picos, menos colas). Mientras que valores negativos indican una distribuci칩n de los datos m치s alejados de la media; es decir, con colas m치s ligeras (menos picos, m치s colas).

Luego, una distribuci칩n normal idealmente tiene una asimetr칤a cercana a 0 y una curtosis cercana a 0.

-----------------
Mediante la **Prueba de Shapiro-Wilks** se mide el grado de ajuste existente entre la distribuci칩n del conjunto de datos analizados y una distribuci칩n normal. El objetivo es validar si los datos provienen de una poblaci칩n tienen la distribuci칩n te칩rica especificada contrastando si las observaciones podr칤an razonablemente proceder de una distribuci칩n gaussiana.

Esta prueba estad칤stica contrasta la hip칩tesis nula con la hip칩tesis alternativa, siendo:
- **Ho:** Los datos provienen de una distribuci칩n normal.
- **Ha:** Los datos no siguen una distribuci칩n normal.

Se eval칰a el nivel de significancia (0,05), tal que:
- si **p-value <= 0,05** se rechaza la hip칩tesis nula, sugiriendo que la distribuci칩n de los datos no es normal.
- si **p-value > 0,05** no se puede rechazar la hip칩tesis nula, sugiriendo que los datos siguen una distribuci칩n normal.

---
**NOTA:** Este m칠todo es adecuado para muestras peque침as y medianas. Las **pruebas de normalidad**, son pruebas poco potentes cuando el tama침o de la muestra es peque침o (n <= 30). Al basarse en la hip칩tesis nula de normalidad, podemos no alcanzar significaci칩n estad칤stica por falta de potencia estad칤stica, asumiendo err칩neamente que los datos siguen una distribuci칩n normal (al no poder rechazar la hip칩tesis nula). Por otra parte, cuando la muestra es muy grande (n > 30), ocurre lo contrario: ser치 suficiente una peque침a desviaci칩n de la normalidad para que la prueba nos d칠 una p significativa y rechacemos la hip칩tesis nula, cuando la mayor parte de las t칠cnicas param칠tricas tolerar칤an peque침as desviaciones de la normalidad si la muestra es grande. Aunque, para estos casos aplica el ***Teorema Central del L칤mite*** que indica que cuando se extraen muestras de tama침o mayor a 30, la distribuci칩n muestral de medias tiene un comportamiento aproximadamente normal.


```python
# Se arma un diciconario con las variables por tipo a analizar con el test estad칤stico
diccionario = {'fixed acidity':1, 'alcohol':1, 'total sulfur dioxide':1, 'pH':1, 'chlorides':1, 'density':0, 'pH':0, 'chlorides':0, 'citric acid':0}
```

```python
# Se aplica el test de Shapiro-Wilks a las variables identificadas

nivel_significancia = 0.05

# Test de Shapiro
for clave, valor in diccionario.items():
  if diccionario[clave] == 1:
      shapiro_test = shapiro(df_white[clave])
      vino = "blanco"
  else:
      shapiro_test = shapiro(df_red[clave])
      vino = "tinto"
  shapiro_pvalue = shapiro_test.pvalue
  distribucion = lambda shapiro_pvalue: 'siguen' if shapiro_pvalue > nivel_significancia else 'NO siguen'
  resultado = distribucion(shapiro_pvalue)
  print(f'El p-value de "{clave}" (para vino {vino}) es {shapiro_pvalue}, por lo tanto los datos {resultado} una distribuci칩n normal.')
```

-----------------
El **QQ-Plot** es un gr치fico que tambi칠n nos permite analizar la normalidad de los datos mediante la comparaci칩n de cuantiles. Permite comparar la distribuci칩n de un conjunto de datos con una distribuci칩n te칩rica, como la distribuci칩n normal.

De esta manera, si los datos siguen una distribuci칩n normal, se alinear치n cerca de la diagonal del gr치fico. Cuanto m치s se alejen, menos probable ser치 que nuestros datos sigan una distribuci칩n normal.

```python
# QQ plot para complementar el test estad칤stico

for clave, valor in diccionario.items():

  if diccionario[clave] == 1:
    sm.qqplot(df_white[clave], line='s')
    vino = "blanco"
  else:
    sm.qqplot(df_red[clave], line='s')
    vino = "tinto"

  plt.xlabel('Cuantiles te칩ricos')
  plt.ylabel('Cuantiles de los datos')
  plt.title(clave + " - " + vino)
  plt.show()
```

- La variable **"chlorides"** para los vinos blancos no sigue la l칤nea recta diagonal de la distribuci칩n normal.

De esta manera, cualquier valor que quede fuera del rango mencionado se considerar치n una anomal칤a.

```python
# Se arma un diciconario con las variables que se aproximan a la distribuci칩n normal
variablesNormales = {'density':0, 'pH':0}
```

```python
# Se calcula el z-score para identificar outliers

totalOutliers = pd.DataFrame()

for clave, valor in variablesNormales.items():

  if valor == 1: # vino = "blanco"
    df_white.loc[:, "variableZ"] = (df_white[clave] - df_white[clave].mean()) / df_white[clave].std()
    df_white.loc[:, "Outlier"] = 0
    df_white.loc[(df_white["variableZ"] > 3) | (df_white["variableZ"] < -3), "Outlier"] = 1
    outliers = df_white.copy()

  else: # vino = "tinto"
    df_red.loc[:, "variableZ"] = (df_red[clave] - df_red[clave].mean()) / df_red[clave].std()
    df_red.loc[:, "Outlier"] = 0
    df_red.loc[(df_red["variableZ"] > 3) | (df_red["variableZ"] < -3), "Outlier"] = 1
    outliers = df_red.copy()

  # Se agrega una columna para identificar la variable
  outliers.loc[:, 'variable'] = clave
  outliers.loc[:, 'type'] = valor

  # se concatenan  los resultados
  totalOutliers = pd.concat([totalOutliers, outliers], ignore_index=True)

resultadoOutliers = totalOutliers.groupby(["variable", "type", "Outlier"]).size().reset_index(name='count')
resultadoOutliers
```

```python
# Se listan los valores outliers

outliersZ = totalOutliers[totalOutliers['Outlier'] == 1]
outliersZ
```

```python
# se calcula la proporci칩n de valores at칤picos respecto al total del dataset bajo estudio.
porcAtipicos = len(outliersZ) / len(df)
print(f"Los valores at칤picos identificados representan el {porcAtipicos:.3f} % del dataset")
```

Dado que la cantidad de registros identificados no es significativa se procede a eliminarlos del conjunto de datos

```python
print("Dimensiones del dataset antes de la remoci칩n de outliers: ", df.shape)
df = df[~df.index.isin(outliersZ.index)]
print("Dimensiones del dataset luego de la remoci칩n de outliers: ", df.shape)
```

### **餃쩚잺 ESCALADO**
**Estandarizaci칩n de los datos**
----

El ***Escalado*** es una parte importante de la Ingenier칤a de Caracter칤sticas (Feature Engineering). Se trata de una t칠cnica que ajusta los valores de las variables para que est칠n dentro de un rango espec칤fico. Esto es crucial en modelos de machine learning ya que muchos algoritmos de machine learning son sensibles a las diferencias de escala entre las variables, pudiendo afectar negativamente el rendimiento del modelo.

Debido a que el dataset con el que se est치 trabajando tiene valores extremos que no ser치n eliminados, utilizaremos el m칠todo ***Robust Scaling*** que utiliza la mediana y el rango intercuart칤lico (IQR) en lugar de la media y la desviaci칩n est치ndar, lo que lo hace m치s robusto a los outliers:

$$ X_{scaled} = \frac{X_i - X_{med}}{X_{q75} - X_{q25}}$$

```python
rs = RobustScaler()
df[['fixed acidity_RS', 'volatile acidity_RS', 'citric acid_RS', 'residual sugar_RS', 'chlorides_RS', 'free sulfur dioxide_RS', 'total sulfur dioxide_RS', 'density_RS', 'pH_RS', 'sulphates_RS', 'alcohol_RS']] = rs.fit_transform(df[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']])
df[['fixed acidity_RS', 'volatile acidity_RS', 'citric acid_RS', 'residual sugar_RS', 'chlorides_RS', 'free sulfur dioxide_RS', 'total sulfur dioxide_RS', 'density_RS', 'pH_RS', 'sulphates_RS', 'alcohol_RS']].describe().round(3)
```
